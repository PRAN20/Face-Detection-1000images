```python
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
for folder in os.listdir("/content/gender/Men/"):
  for imgs in os.listdir("/content/gender/Men/"+folder):
    img= cv2.imread("/content/gender/Men/"+folder+"/"+imgs,1)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    faceCascade = cv2.CascadeClassifier('haarcascade_frontalface_alt2.xml')

    faces = faceCascade.detectMultiScale(
    gray, scaleFactor = 1.1, minNeighbors = 5, minSize = (30,30)
    )

    image = cv2.imread("/content/gender/Men/"+folder+"/"+imgs,0)    
    for (x, y, w, h) in faces:
      cv2.rectangle(image, (x, y), (x+w, y+h), (255,0,0,), 2)

    ims = []
    for (x, y, w, h) in faces:
      crop = image[y:y+h, x:x+w]
      ims.append(crop)
    cv2.imwrite("cropped/" + imgs, crop)
```

```python
from zipfile import ZipFile
import os
  
def get_all_file_paths(directory):
  
    # initializing empty file paths list
    file_paths = []
  
    # crawling through directory and subdirectories
    for root, directories, files in os.walk(directory):
        for filename in files:
            # join the two strings in order to form the full filepath.
            filepath = os.path.join(root, filename)
            file_paths.append(filepath)
  
    # returning all file paths
    return file_paths        
  
def main():
    # path to folder which needs to be zipped
    directory = './cropped'
  
    # calling function to get all file paths in the directory
    file_paths = get_all_file_paths(directory)
  
    # printing the list of all files to be zipped
    print('Following files will be zipped:')
    for file_name in file_paths:
        print(file_name)
  
    # writing files to a zipfile
    with ZipFile('my_python_files.zip','w') as zip:
        # writing each file one by one
        for file in file_paths:
            zip.write(file)
  
    print('All files zipped successfully!')        
  
  
if __name__ == "__main__":
    main()
```
